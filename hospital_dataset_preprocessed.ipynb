```python
import numpy as np

bill = np.array([5000, 15000, 85000, 40000, 18000])

print("Hospital Bill Data:", bill)
print("Total Bill Amount:", np.sum(bill))
print("Average Bill Amount:", np.mean(bill))
print("Highest Bill Amount:", np.max(bill))
print("Lowest Bill Amount:", np.min(bill))
print("Sorted Bill Amount:", np.sort(bill))
```

```python
import pandas as pd

data = { "Patient": ["Arun", "Priya", "Ravi", "Divya", "Karthik"],
         "Age": [25, 30, 45, 28, 35],
         "Bill_Amount": [5000, 15000, 85000, 40000, 18000]}

df = pd.DataFrame(data)

print(df.head())
print(df.info())
print(df.describe())
print(df.sort_values("Bill_Amount"))
print(df["Bill_Amount"].mean())
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    "Patient": ["Arun", "Priya", "Ravi", "Divya", "Karthik"],
    "Bill": [5000, 15000, 85000, 40000, 18000]}

df = pd.DataFrame(data)

sns.barplot(x="Patient", y="Bill", data=df)
plt.show()
sns.lineplot(x="Patient", y="Bill", data=df)
plt.show()
sns.scatterplot(x="Patient", y="Bill", data=df)
plt.show()
sns.histplot(df["Bill"])
plt.show()
sns.boxplot(y="Bill", data=df)
plt.show()
```

```python
import matplotlib.pyplot as plt

patient = ["Arun", "Priya", "Ravi", "Divya", "Karthik"]
bill = [5000, 15000, 85000, 40000, 18000]

plt.plot(patient, bill)
plt.show()

plt.bar(patient, bill)
plt.show()

plt.scatter(patient, bill)
plt.show()

plt.pie(bill, labels=patient, autopct="%1.1f%%")
plt.show()

plt.hist(bill)
plt.show()
```
